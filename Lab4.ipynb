```python
#Part A: Data Processing
import numpy as np

# Given data
data = [120, -50, 200, 0, 75, -10, 300]

# 1. Removing negative values
filtered_data = list(filter(lambda x: x >= 0, data))

# 2. Increasing each value by 10%
increased_data = list(map(lambda x: x * 1.10, filtered_data))

# 3. Converting to NumPy array
numpy_array = np.array(increased_data)

print("Processed NumPy Array:")
print(numpy_array)
```

```python
#Part B: Build a Simple Neural Network
import numpy as np

# Input
X = np.array([[100, 2],
              [200, 3],
              [150, 1]])

# Weights
W = np.array([[0.01, 0.5],
              [0.02, 0.3]])

# Bias
b = np.array([0.1, 0.2])

# 1. Compute z = XW + b
z = np.dot(X, W) + b

# 2. Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# 3. Applying sigmoid
output = sigmoid(z)

# 4. Printing output
print("\nZ (Linear Combination):")
print(z)

print("\nFinal Output after Sigmoid:")
print(output)
```

```python
#Part C: Combine Functional Programming + Neural Network

# Using processed data from Part A
# Taking first 4 values
subset = numpy_array[:4]

# Reshaping into 2x2
X_new = subset.reshape(2, 2)

print("\nNew Input (2x2):")
print(X_new)

# Passing through neural network
z_new = np.dot(X_new, W) + b
output_new = sigmoid(z_new)

print("\nZ (New Data):")
print(z_new)

print("\nFinal Output (New Data):")
print(output_new)
```
